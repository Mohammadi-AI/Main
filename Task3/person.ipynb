```python
class person:
    
    firstname=[]
    lastname=[]
    age=[]
    
    def __init__ (self,firstname, lastname, age):
       
        self.firstname+=firstname
        self.lastname+=lastname
        self.age+=age
        
    def person_sort(self)    :
        self.firstname.sort()
        self.lastname.sort()
        self.age.sort()
        
    
p1_firstname=input("please enter your firstname : ")
p1_lastname= input("please enter your lastname : ")  
p1_age=input("please enter your age :")

p1=person(p1_firstname,p1_lastname,p1_age)
p1.person_sort()

print()

p2_firstname=input("please enter your firstname : ")
p2_lastname= input("please enter your lastname : ")  
p2_age=input("please enter your age :")

p2=person(p2_firstname,p2_lastname,p2_age)
p2.person_sort()

print()

p3_firstname=input("please enter your firstname : ")
p3_lastname= input("please enter your lastname : ")  
p3_age=input("please enter your age :")

p3=person(p3_firstname,p3_lastname,p3_age)
p3.person_sort()

finallist_name=""
finallist_family=""
finallist_age=""


for i in range(3):
    if(person.firstname[i])==(p1_firstname):
        finallist_name+="p1"
        finallist_name+=" "
    elif(person.firstname[i])==(p2_firstname):
        finallist_name+="p2"
        finallist_name+=" "
    elif (person.firstname[i])==(p3_firstname):
        finallist_name+="p3"
        finallist_name+=" "
    
    if(person.lastname[i])==(p1_lastname):
        finallist_family+="p1"
        finallist_family+=" "
    elif(person.lastname[i])==(p2_lastname):
        finallist_family+="p2"
        finallist_family+=" "
    elif (person.lastname[i])==(p3_lastname):
        finallist_family+="p3"
        finallist_family+=" "
    
    if(person.age[i])==(p1_age):
        finallist_age+="p1"
        finallist_age+=" "
    elif(person.age[i])==(p2_age):
        finallist_age+="p2"
        finallist_age+=" "
    elif (person.age[i])==(p3_age):
        finallist_age+="p3"
        finallist_age+=" "
    
    
print("sort by firstname :  ",finallist_name)    
print("sort by lastname :  ",finallist_family)
print("sort by age :  ",finallist_age)

```
